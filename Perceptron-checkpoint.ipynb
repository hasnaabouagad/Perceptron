```python
import numpy as np
import matplotlib.pyplot as plt 

#preparation des donnees lineairement separables 
(fig,ax)=plt.subplots()

p1 = np.array([np.random.uniform(-1, 1), np.random.uniform(-1, 1)])
p2 = np.array([np.random.uniform(-1, 1), np.random.uniform(-1, 1)])
point1=np.array([p1[0],p2[0]])
point2=np.array([p1[1],p2[1]])
plt.plot(point1,point2,color='black')
is_above = lambda p,p1,p2: np.cross(p-p1, p2-p1) < 0
N=100
newtr=[]
training_points=[]
training_points_y_1=[]
training_points_y_0=[]
training_y=[]
plt.axis([-1, 1, -1, 1])
for i in range(2*N):
    p=[np.random.uniform(-1, 1), np.random.uniform(-1, 1)]
    if is_above(p,p1,p2):
        ax.scatter(p[0], p[1], color='purple')
        training_points.append(p)
        training_points_y_1.append(p)
        training_y.append(1)
    
            
    else:
        ax.scatter(p[0], p[1], color='deeppink')
        training_points.append(p)
        training_points_y_0.append(p)
        training_y.append(-1)
plt.show()
training_points=np.array(training_points)
training_y=np.array(training_y)
newtr=np.transpose(training_points)
training_points_y_1=np.array(training_points_y_1)
training_points_y_0=np.array(training_points_y_0)

```

```python
class Perceptron:
    
    def initial(self,training_points,training_y,nb_iterations):
        self.n_samples=np.shape(training_points)[0] 
        self.n_features=np.shape(training_points)[1] 
        self.weight=np.zeros((self.n_features+1,))
        self.newtr=np.concatenate((np.ones((1,self.n_samples)),newtr),axis=0)
        
       
        for i in range(nb_iterations):
            for j in range(self.n_samples):
                if training_y[j]*(self.weight.dot(self.newtr[:,j])) <=0:
                    self.weight +=training_y[j]*self.newtr[:,j]
    def droite(self):
        (fig,ax1)=plt.subplots()
        plt.axis([-1, 1, -1, 1])
        x1 = np.amin(training_points[:,0])
        x2 = np.amax(training_points[:,0])
        x3 = (-self.weight[0] - self.weight[1]* x1 ) / self.weight[2]
        x4 = (-self.weight[0] - self.weight[1]* x2 ) / self.weight[2]

        plt.plot([x1, x2],[x3, x4],color='blue')
        plt.plot(point1,point2,color='black')
        ax1.scatter(training_points_y_1[:,0],training_points_y_1[:,1],color='red')
        ax1.scatter(training_points_y_0[:,0],training_points_y_0[:,1],color='green')
        plt.show()
```

```python
p=Perceptron()
```

```python
p.initial(training_points,training_y,1000)
```

```python
p.droite()
```
